```python
# Exercise 1 - Turn the shopping cart program from yesterday into an object-oriented program
# The comments in the cell below are there as a guide for thinking about the problem. However, if you feel a different way is best for you and your own thought process, please do what feels best for you by all means.

# Create a class called cart that retains items and has methods to add, remove, and show

shopping_list = ["beef", "chicken", "fish", "pork", "apple", "banana"]

class main():
    def costco():
        print()
        for i in shopping_list:
                print(i)

    def additem():
        item = input("Add to list: ")
        shopping_list.append(item)
        print(item + " added")

    def removeitem():
        item = input("Which item from the list: ")
        shopping_list.remove(item)
        print(item + " has been removed.")
        
    def check_list():
        item = input("what would you like to check on the shopping list: ")
        if item in shopping_list:
            print("Yes, " + (item) + " is on the shopping list")
        else:
            print("No, " + (item) + " is not on the shopping list")
        
    def quit():
        for i in shopping_list:
            print(i)
    
    while True:
        print(''' ### SHOPPING CART ###
        1. View Original List
        2. Add Item
        3. Delete Item
        4. check items on the list
        ''')
        
        selection = input("input#: ")

        if selection == "1":
            costco()
        elif selection == "2":
            additem()
        elif selection == "3":
            removeitem()
        elif selection == "4":
            check_list()
        elif selection == "quit":
            print(shopping_list)
            break

main()


# # Exercise 2 - Create a class 'Ford' that inherits from 'Car' class and initialize it as a 
# Blue Ford Explorer with 4 wheels using the super() method, add a method to car which prints 'this is a car' 
# and create an overriding method in class Ford that prints the information on the car
# Create Parent Class called Car - That will be inherited by child Class called Ford
# Add a method to the Car class that prints "This is a car"
# Inside of the Child Class initialize the the inherited info using super()
# Override the above method with the information of the Ford Car

class Car:
    def __init__(self, color, car, name):
        self.color = blue
        self.car = ford
        self.name = explorer
    
    def explorer(self):
        print("\nThis is a car: +self.color, self.car, self.name")
        
class Ford(Car):
    super().init(self, color, car, name)
    self.car = []


```
